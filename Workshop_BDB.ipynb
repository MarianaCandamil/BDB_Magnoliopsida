# Transposable Elements Database

```python
df = pd.read_csv('/content/drive/MyDrive/MBBC/Segundo Semestre/BDB/ptg000002l.fasta.TEanno.gff3', sep='\t')
```

```python
print(df)
```

```python
sequences=SeqIO.parse("ptg000002l.fasta","fasta")
lista=[]
for seq in sequences:
  for i in range(len(df)):
    inicio=df.iloc[i,3]
    final=df.iloc[i,4]
    lista.append(str(seq.seq[inicio:final]))
```

```python
df2=df
```

Adding the information of the sequence to the dataframe

```python
df2=df2.assign(sequence=lista)
```

```python
df2
```

```python
#Save the información
file1=open("ptg000002l_format.json","w")
cromosomas=df2.to_json(orient="records")
with open("/content/drive/MyDrive/MBBC/Segundo Semestre/BDB/ptg000002l_format.json","w") as output:
  output.write(cromosomas)
```

# Predicted Genes

```python
df_genes = pd.read_csv('/content/drive/MyDrive/MBBC/Segundo Semestre/BDB/augustus.abinitio.pfam_results.csv', sep=',')
```

```python
print(df_genes)
```

```python
df_blast = pd.read_csv('/content/drive/MyDrive/MBBC/Segundo Semestre/BDB/Blast_NR.abinitio_prot_Dm.tab', sep='\t')
```

```python
print(df_blast)
```

```python
lista_final=[]
for i in range(len(df_genes)):
  lista=[]
  print(i)
  for j in range(len(df_blast)):
    if(str(df_genes.iloc[i,0])==str(df_blast.iloc[j,0])):
      lista.append(str(df_blast.iloc[j,1]))
  lista_final.append(lista)
```

```python
 lista_final
```

```python
df_cat=df_genes
```

```python
#Adding the reference accesion of BLAST
df_cat=df_cat.assign(blast_accesion=lista_final)
```

```python
df_cat
```

```python
sequences=SeqIO.parse("augustus.abinitio.aa","fasta")
lista_sequences=[]
for seq in sequences:
  for i in range(len(df_cat)):
    if(str(seq.id)==str(df_cat.iloc[i,0])):
      lista_sequences.append(str(seq.seq))
```

```python
#Adding the gene sequence:
df_cat=df_cat.assign(sequences=lista_sequences)
```

```python
df_cat
```

```python
file1=open("genes_format.json","w")
cromosomas=df_cat.to_json(orient="records")
with open("/content/drive/MyDrive/MBBC/Segundo Semestre/BDB/genes_format.json","w") as output:
  output.write(cromosomas)
```
